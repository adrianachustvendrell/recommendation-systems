## 2. Usuarios (`info_usuarios.csv`, `prefs_usuarios.csv`)

### 2.1 Usuarios datos

```python
import pandas as pd

with open("rawdata/usuarios_datos_personales.txt", "r", encoding="latin1") as file:
    usuarios_datos_content = file.readlines()

# Extracting valid user records by enforcing strict 8-field grouping (5 known + 3 additional fields)
cleaned_records = []
current_record = []

for line in usuarios_datos_content:
    line = line.strip()

    if len(current_record) == 0 and line.isdigit():  # First field: id_usuario
        current_record.append(line)
    elif len(current_record) == 1 and not line.isdigit():  # Second field: nombre_usuario (non-numeric)
        current_record.append(line)
    elif len(current_record) == 2 and line.isdigit():  # Third field: edad (numeric)
        current_record.append(line)
    elif len(current_record) == 3 and not line.isdigit():  # Fourth field: sexo (non-numeric, usually 'F' or 'M')
        current_record.append(line)
    elif len(current_record) == 4 and line.isdigit():  # Fifth field: ocupacion_id (numeric)
        current_record.append(line)
    elif len(current_record) in [5, 6, 7] and line.isdigit():  # Next three fields: Hijos, Edad hijo menor, Edad hijo mayor
        current_record.append(line)

    if len(current_record) == 8:  # Once we have all required fields, save the record
        cleaned_records.append(current_record)
        current_record = []

# Creating a DataFrame with correctly structured user records
usuarios_datos = pd.DataFrame(cleaned_records, columns=[
    "id_usuario", "nombre_usuario", "edad", "sexo", "id_ocupacion",
    "hijos", "edad_hijo_menor", "edad_hijo_mayor"
])

# Convert data types
usuarios_datos["id_usuario"] = usuarios_datos["id_usuario"].astype(int)
usuarios_datos["edad"] = usuarios_datos["edad"].astype(int)
usuarios_datos["id_ocupacion"] = usuarios_datos["id_ocupacion"].astype(int)
usuarios_datos["hijos"] = usuarios_datos["hijos"].astype(int)
usuarios_datos["edad_hijo_menor"] = usuarios_datos["edad_hijo_menor"].astype(int)
usuarios_datos["edad_hijo_mayor"] = usuarios_datos["edad_hijo_mayor"].astype(int)
```

```python
with open("rawdata/ocupaciones.txt", "r", encoding="latin1") as file:
    ocupaciones_content = file.readlines()

ocupacion_ids = ocupaciones_content[0::2]
ocupacion_names = ocupaciones_content[1::2]

ocupaciones = pd.DataFrame({
    "id_ocupacion": [int(x.strip()) for x in ocupacion_ids],
    "ocupacion": [x.strip() for x in ocupacion_names]})

usuarios_datos = usuarios_datos.merge(ocupaciones, on="id_ocupacion", how="left")
usuarios_datos.head()
```

```python
tipos_usuarios = []

for index, row in usuarios_datos.iterrows():
    if row['edad'] < 60:
        #tipo 2: viajero borrachera
        if row['edad'] > 18 and (row['ocupacion'] == 'Trabajadores no cualificados' or row['ocupacion'] == 'Inactivo o desocupado') and row['hijos'] == 0:
            tipos_usuarios.append('tipo2')
        
        #tipo 5: viajero niños
        elif row['edad'] > 30 and row['hijos'] > 0 and (row['edad_hijo_menor'] < 12 or row['edad_hijo_mayor'] < 12):
            tipos_usuarios.append('tipo5')
        
        #tipo 3: viajero joven con cultura 
        elif row['edad'] < 30 and row['edad'] > 18 and (row['ocupacion'] == 'Dirección de las empresas y de las administraciones públicas' or row['ocupacion'] == 'Técnicos y profesionales científicos e intelectuales'or row['ocupacion'] == 'Técnicos y profesionales de apoyo'):
            tipos_usuarios.append('tipo3')
        
        else:
            #tipo 6: 
            if row['ocupacion'] == 'Fuerzas armadas':
                tipos_usuarios.append('tipo6')

            #tipo 1: viajero disfrute (grupo grande)  
            else:
             tipos_usuarios.append('tipo1')
    
    #tipo 4: viajero jubilado
    else:
        tipos_usuarios.append('tipo4')
```

```python
usuarios_datos['tipos_usuario'] = tipos_usuarios
usuarios_datos = usuarios_datos[usuarios_datos['id_usuario'] != 181]
usuarios_datos.tail()
```

```python
usuarios_datos.to_csv("./data/info_usuarios.csv", index=False, encoding="utf-8")
```

### 2.2 Usuarios preferencias

```python
with open("rawdata/usuarios_preferencias.txt", "r", encoding="latin1") as file:
    usuarios_prefs_content = file.readlines()

usuarios_prefs_records = [usuarios_prefs_content[i:i+3] for i in range(0, len(usuarios_prefs_content), 3)]
usuarios_prefs = pd.DataFrame(usuarios_prefs_records, columns=["id_usuario", "id_categoria", "calificacion"])

usuarios_prefs = usuarios_prefs.map(lambda x: x.strip() if isinstance(x, str) else x)
usuarios_prefs["id_usuario"] = usuarios_prefs["id_usuario"].astype(int)
usuarios_prefs["id_categoria"] = usuarios_prefs["id_categoria"].astype(int)
usuarios_prefs["calificacion"] = usuarios_prefs["calificacion"].astype(int) * 10  # Convert scale 1-10 to 1-100


with open("rawdata/preferencias.txt", "r", encoding="latin1") as file:
    categorias_content = file.readlines()

categoria_ids = []
categoria_names = []

for i in range(0, len(categorias_content), 3):  # Every 3 lines, extract relevant data
    if i + 1 < len(categorias_content):  # Ensure we don't go out of bounds
        categoria_ids.append(int(categorias_content[i].strip()))
        categoria_names.append(categorias_content[i + 1].strip())

# Creating corrected preferences DataFrame
categorias = pd.DataFrame({
    "id_categoria": categoria_ids,
    "categoria": categoria_names})

usuarios_prefs_final = usuarios_prefs.merge(categorias, on="id_categoria", how="left")
usuarios_prefs_final
```

```python
usuarios_prefs_final.to_csv("./data/prefs_usuarios.csv", index=False, encoding="utf-8")
```

```python
with open("rawdata/puntuaciones_usuario_base.txt", "r", encoding="latin-1") as file:
    # Leemos todas las líneas del archivo
    lines = file.readlines()

# Limpiamos las líneas (removemos saltos de línea)
lines = [line.strip() for line in lines]

# Dividimos las líneas en grupos de 3
data = []
for i in range(0, len(lines), 3):
    id = lines[i]
    lugar = lines[i+1]
    ratio = lines[i+2]
    data.append([id, lugar, ratio])

# Convertimos a DataFrame
df_base = pd.DataFrame(data, columns=['id_usuario', 'id_item', 'ratio'])
df_base.head()
```

```python
with open("rawdata/puntuaciones_usuario_test.txt", "r", encoding="latin-1") as file:
    # Leemos todas las líneas del archivo
    lines = file.readlines()

# Limpiamos las líneas (removemos saltos de línea)
lines = [line.strip() for line in lines]

# Dividimos las líneas en grupos de 3
data = []
for i in range(0, len(lines), 3):
    id = lines[i]
    lugar = lines[i+1]
    ratio = lines[i+2]
    data.append([id, lugar, ratio])

# Convertimos a DataFrame
df_test = pd.DataFrame(data, columns=['id_usuario', 'id_item', 'ratio'])
df_test.head()
```

```python
# Convertir las columnas 'ratio' a tipo numérico
df_base['ratio'] = pd.to_numeric(df_base['ratio'])  
df_test['ratio'] = pd.to_numeric(df_test['ratio'])

# Normalizar y multiplicar por 100
df_base['ratio'] = (df_base['ratio'] - df_base['ratio'].min()) / (df_base['ratio'].max() - df_base['ratio'].min()) * 100
df_test['ratio'] = (df_test['ratio'] - df_test['ratio'].min()) / (df_test['ratio'].max() - df_test['ratio'].min()) * 100

df_base['ratio'] = df_base['ratio'].round(4)
df_test['ratio'] = df_test['ratio'].round(4)
```
